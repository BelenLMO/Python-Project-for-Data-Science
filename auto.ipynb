## Normalización de datos
¿Por qué la normalización?

La normalización es el proceso de transformar los valores de varias variables en un rango similar. Las normalizaciones típicas incluyen

escalar la variable de modo que el promedio de la variable sea 0
escalar la variable de modo que la varianza sea 1
escalar la variable de modo que los valores de la variable oscilen entre 0 y 1
Ejemplo

Para demostrar la normalización, digamos que desea escalar las columnas "longitud", "ancho" y "altura".

Objetivo: normalizar esas variables de modo que sus valores oscilen entre 0 y 1

Enfoque: reemplazar el valor original por (valor original)/(valor máximo)

Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include

scaling the variable so the variable average is 0
scaling the variable so the variance is 1
scaling the variable so the variable values range from 0 to 1
Example

To demonstrate normalization, say you want to scale the columns "length", "width" and "height".

Target: normalize those variables so their value ranges from 0 to 1

Approach: replace the original value by (original value)/(maximum value)

## Binning
¿Por qué binning?

Binning es un proceso de transformación de variables numéricas continuas en 'bins' categóricos discretos para el análisis agrupado.

Ejemplo:

En su conjunto de datos, "caballos de fuerza" es una variable con un valor real que va de 48 a 288 y tiene 59 valores únicos. ¿Qué sucede si solo le interesa la diferencia de precio entre los autos con alta potencia, potencia media y poca potencia (3 tipos)? Puede reorganizarlos en tres 'bins' para simplificar el análisis.

Utilice el método Pandas 'cut' para segmentar la columna 'caballos de fuerza' en 3 bins.

Ejemplo de binning de datos en Pandas
Convierta los datos al formato correcto:

```python
df.dtypes
```

```python
# Verificar los elementos únicos y su frecuencia en la columna "horsepower"
horsepower_values = df["horsepower"].value_counts(dropna=False)

# Mostrar los resultados
print(horsepower_values)
```

```python
# Paso 1: Identificar la moda de la columna "horsepower"
mode_horsepower = df["horsepower"].mode()[0]

# Paso 2: Reemplazar valores que superen el máximo permitido (288) con la moda
df.loc[df["horsepower"] > 288, "horsepower"] = mode_horsepower

# Verificar si el reemplazo fue exitoso
print(df["horsepower"].value_counts(dropna=False))
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

Grafique el histograma de caballos de fuerza para ver la distribución de caballos de fuerza.

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

Encuentre 3 contenedores con el mismo ancho de banda mediante la función linspace(start_value, end_value, numbers_generated de Numpy.

Como desea incluir el valor mínimo de caballos de fuerza, establezca start_value = min(df["horsepower"]).

Como desea incluir el valor máximo de caballos de fuerza, establezca end_value = max(df["horsepower"]).

Como está construyendo 3 contenedores de igual longitud, necesita 4 divisores, por lo que numbers_generated = 4.

Construya una matriz de contenedores con un valor mínimo a un valor máximo utilizando el ancho de banda calculado anteriormente. Los valores determinarán cuándo termina un contenedor y comienza otro.

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(29)
```

```python
df['horsepower-binned'].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

## Indicator Variable
<b>What is an indicator variable?</b>
<p>
    An indicator variable (or dummy variable) is a numerical variable used to label categories. They are called 'dummies' because the numbers themselves don't have inherent meaning. 

    Una variable indicadora (o variable ficticia) es una variable numérica que se utiliza para etiquetar categorías. Se denominan "variables ficticias" porque los números en sí mismos no tienen un significado inherente.
</p>

<b>Why use indicator variables?</b>
<p>
    You use indicator variables so you can use categorical variables for regression analysis in the later modules.

    Las variables indicadoras se utilizan para poder utilizar variables categóricas para el análisis de regresión en los módulos posteriores.
</p>
<b>Example</b>
<p>
    The column "fuel-type" has two unique values: "gas" or "diesel". Regression doesn't understand words, only numbers. To use this attribute in regression analysis, you can convert "fuel-type" to indicator variables.

    La columna "fuel-type" tiene dos valores únicos: "gas" o "diesel". La regresión no entiende palabras, solo números. Para utilizar este atributo en el análisis de regresión, puede convertir "fuel-type" en variables indicadoras.
</p>

<p>
    Use the Panda method 'get_dummies' to assign numerical values to different categories of fuel type. 

    Utilice el método Panda 'get_dummies' para asignar valores numéricos a diferentes categorías de tipo de combustible.
</p>

Obtenga las variables indicadoras y asígnelas al marco de datos "dummy_variable_1":

```python
# Verificar los elementos únicos y su frecuencia en la columna "hfuel-type"
fuel = df["fuel-type"].value_counts(dropna=False)

# Mostrar los resultados
print(fuel)
```

```python
# Paso 1: Convertir todos los valores de "fuel-type" a minúsculas
df["fuel-type"] = df["fuel-type"].str.lower().str.strip()

# Paso 2: Reemplazar variantes de "gas" y "diesel" con los valores correctos
df["fuel-type"] = df["fuel-type"].replace({
    "gas": "gas",
    "gasolina": "gas",
    "diésel": "diesel"
})

# Verificar el resultado
print(df["fuel-type"].value_counts())
```

```python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)
```
